# Exercícios de Matplotlib

## 1. Crie um gráfico de linha para a função y = x3 no intervalo de x = −10 até x = 10

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**3

plt.plot(x, y)
plt.title('Gráfico de linha para função y = xˆ3')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

## 2. Plote um gráfico de dispersão para 100 pontos aleatórios com uma paleta de cores baseada nos valores de y

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, c=y, cmap='viridis')
plt.title('Gráfico de dispersão com paleta de cores')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.colorbar(label='Valor de y')
plt.show()
```

## 3. Gere um histograma para 500 números inteiros gerados aleatoriamente entre 0 e 100

```python
dados = np.random.randint(0, 101, 500)

plt.hist(dados, bins=30, edgecolor='black')
plt.title('Histograma de 500 números aleatorios entre 0 e 100')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

## 4. Crie um gráfico de barras para 5 categorias com valores diferentes, colorindo cada barra de uma cor diferente

```python
categorias = ['A','B','C','D','E']
valores = [43, 2, 54, 34, 9]

plt.bar(categorias, valores, color=['blue', 'red', 'purple', 'yellow', 'green'])
plt.title('Gráfico de Barras')
plt.xlabel('Categoria')
plt.ylabel('Valores')
plt.show()
```

## 5. Faça um gráfico de pizza para representar as porcentagens de quatro setores de mercado com rótulos personalizados

```python
labels = ['A','B','C','D']
sizes = [30, 45, 5, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Gráfico de Pizza')
plt.show()
```

## 6. Plote dois gráficos de linha em subplots diferentes: y1 = ex e y2 = log(x)

```python
x = np.linspace(1, 10, 100)

y1 = np.exp(x)
y2 = np.log(x)

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Gráfico de linha y1 = e^x')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Gráfico de linha y2 = log(x)')

plt.tight_layout()
plt.show()
```

## 7. Crie um gráfico de dispersão com tamanho dos pontos variando de acordo com os valores de y

```python
x = np.random.rand(100)
y = np.random.rand(100)

sizes = y * 100

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title('Gráfico de dispersão com tamanho variável')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

## 8. Plote um gráfico de linha com intervalos de confiança (sombramento) para a função y = sin(x)

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Função y = sin(x)')
plt.fill_between(x, y - 0.1, y + 0.1, color='blue', alpha=0.2)
plt.title('Gráfico de linha com intervalos de confiança')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()
plt.show()
```

## 9. Adicione título, rótulos e uma grade personalizada a um gráfico de linha

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.plot(x, y, color='blue', linestyle='--', marker='o')
plt.title('Gráfico de linha com título e grade personalizada')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.grid(True)
plt.show()
```

## 10. Plote três gráficos de linha em um único gráfico, cada um com uma cor e estilo de linha diferentes

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.subplot(3, 1, 1)
plt.plot(x, y1, color='blue', linestyle='-')
plt.title('Gráfico Seno')

plt.subplot(3, 1, 2)
plt.plot(x, y2, color='red', linestyle='--')
plt.title('Gráfico Cosseno')

plt.subplot(3, 1, 3)
plt.plot(x, y3, color='green', linestyle=':')
plt.title('Grafico Tangente')

plt.tight_layout()
plt.show()
```
